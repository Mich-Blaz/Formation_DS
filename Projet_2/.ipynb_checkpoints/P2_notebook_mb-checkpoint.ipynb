fonctions utiles plus tard

```python
def print1info(a):
    print('Il y a ',a[0],' redondance dans le dataframe, avec ', a[1],' valeurs manquantes soit un taux de valeurs manquantes de ',a[2],'%. Il y a ', a[3], 'lignes et ', a[4],' colonnes (attributs).') 
def premierinfoDF(df):
    return df.duplicated().sum(), df.isnull().sum().sum(), int(df.isnull().mean().mean()*100),len(df.index),len(df.columns)
```

```python
def graph_arraigne(propriete,valeurs,valeursMax=200):
    fig=plt.figure(figsize=(10,8),facecolor='white')
    axes=plt.subplot(111,polar=True)
    t=np.arange(0,np.pi*2,np.pi*2/len(propriete))
    plt.xticks(t,[])
    plt.yticks(np.linspace(0,valeursMax,15))
    points = [(x,y) for x,y in zip(t,valeurs)]
    points.append(points[0])
    points = np.array(points)
    codes = [path.Path.MOVETO,] + \
            [path.Path.LINETO,]*(len(valeurs) -1) + \
            [ path.Path.CLOSEPOLY ]
    _path = path.Path(points, codes)
    _patch = patches.PathPatch(_path, fill=True, color='blue', linewidth=0, alpha=.1)
    axes.add_patch(_patch)
    _patch = patches.PathPatch(_path, fill=False, linewidth = 2)
    axes.add_patch(_patch)


    # Draw circles at value points
    plt.scatter(points[:,0],points[:,1], linewidth=2,
                s=50, color='white', edgecolor='black', zorder=10)

    # Set axes limits
    plt.ylim(0,50)


    # Draw ytick labels to make sure they fit properly
    for i in range(len(propriete)):
        angle_rad = i/float(len(propriete))*2*np.pi
        angle_deg = i/float(len(propriete))*360
        ha = "right"
        if angle_rad < np.pi/2 or angle_rad > 3*np.pi/2: ha = "left"
        plt.text(angle_rad, 10.75, propriete[i], size=14,
                 horizontalalignment=ha, verticalalignment="center")
    plt.show()
```

```python
def comparaison(df_,evolution=False,nom='nom indicateur'):
    
    if evolution==False:
        ser_=df_.mean().sort_values(ascending=False)
        ser=pd.Series(np.arange(1,1+len(ser_)),index=ser_.index)
        return ser, nom
    else:
        return evolution_5_15(df_), nom
def evolution_5_15(df):
    x1=df.loc['2005',:]
    x2=df.loc['2015',:]
    ser_=((x2-x1)/10).sort_values(ascending=False)
    ser=pd.Series(np.arange(1,1+len(ser_)),index=ser_.index)
    return ser

    
    
```
